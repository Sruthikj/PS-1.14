SOBEL EDGE DETECTION

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("/content/taylor.jpg")  # Replace with your image filename
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Sobel filter
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges

# Combine the two directions
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Normalize the result for visualization
sobel_combined = cv2.convertScaleAbs(sobel_combined)

# Display the original and processed images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Sobel Edge Detection
plt.subplot(1, 2, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()
```

PREWITT EDGE DETECTION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("/content/taylor.jpg")  # Replace with your image filename
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define Prewitt kernels
kernel_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
kernel_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

# Apply Prewitt filters
prewitt_x = cv2.filter2D(gray, -1, kernel_x)
prewitt_y = cv2.filter2D(gray, -1, kernel_y)

# Combine the two directions
prewitt_combined = cv2.magnitude(prewitt_x.astype(float), prewitt_y.astype(float))

# Normalize the result for visualization
prewitt_combined = cv2.convertScaleAbs(prewitt_combined)

# Display the original and processed images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Prewitt Edge Detection
plt.subplot(1, 2, 2)
plt.imshow(prewitt_combined, cmap='gray')
plt.title("Prewitt Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()
```
